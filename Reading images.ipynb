### PIL : Python Imaging Library

- Pillow is an image manipulation and processing library
- We can use pillow to crop, resize images and to do basic filtering.
- For advanced tasks that require computer vision or machine learning we have other packages,such as openCV, scikit image and  scikit learn. 

```python
from PIL import Image 
import numpy as np   #numpy to convert images to arrays
```

```python
# Read image 
img = Image.open("images/test_image.jpg") #Not a numpy array
print(type(img))
```

```python
# Output Image in separate window
img.show() 
```

```python
# Output Image in notebook
display(img)
```

```python
# prints format of image 
print(img.format) 
```

```python
# prints mode of image 
print(img.mode) 
```

```python
#PIL image is not by default numpy array but can convert PIL image to numpy array. 
img1 = np.asarray(img)
print(type(img1))
```
